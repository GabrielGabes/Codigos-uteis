# Códigos de Controle de erros:

```python
## Obter o nome do erro

try:
    raise ValueError("Erro de exemplo") # Código que pode causar exceção
except Exception as e:
    print("Error:", e)
```

```python
# Obter Mais Informações com traceback
## O módulo traceback permite obter detalhes sobre onde a exceção ocorreu no código.

import traceback

try:
    raise ValueError("Erro de exemplo") # Código que pode causar exceção
except Exception as e:
    print("Error:", e)
    print("Traceback:")
    print(traceback.format_exc())
```

```python
## Obter o Tipo de Exceção

try:
    raise ValueError("Erro de exemplo") # Código que pode causar exceção
except Exception as e:
    print(f"Error: {e}")
    print(f"Exception Type: {type(e).__name__}")
```

```python
# Incluir Variáveis Locais
## Capturar variáveis locais no momento em que a exceção ocorreu

import sys
import traceback

try:
    x = 10
    y = 0
    result = x / y
except Exception as e:
    print(f"Error: {e}")
    print(f"Exception Type: {type(e).__name__}")
    _, _, tb = sys.exc_info()
    frame = traceback.extract_tb(tb)[-1]
    print(f"File: {frame.filename}")
    print(f"Line: {frame.lineno}")
    print(f"Function: {frame.name}")
    print(f"Local Variables: {frame.line}")
```

```python
## Logging para Depuração Melhorada

import logging
import traceback

logging.basicConfig(filename='error.log', level=logging.ERROR)

try:
    raise ValueError("Erro de exemplo")
except Exception as e:
    logging.error("Exception occurred", exc_info=True)
    print(f"Error: {e}")
    print("Veja os detalhes no arquivo 'error.log'")
```

```python
## Usar o Decorator para Capturar Exceções Globalmente

import logging
import traceback

logging.basicConfig(filename='error.log', level=logging.ERROR)

def exception_logger(func):
    def wrapper(*args, **kwargs):
        try:
            return func(*args, **kwargs)
        except Exception as e:
            logging.error("Exception in %s", func.__name__, exc_info=True)
            print(f"Error: {e}")
            print("Veja os detalhes no arquivo 'error.log'")
    return wrapper

exception_logger
def example_function():
    raise ValueError("Erro de exemplo")

example_function()
```

```python
# Manipulação de Strings
from unidecode import unidecode
#unidecode(string)

# Regex
import re 

# 1. Compila um padrão de expressão regular em um objeto de padrão, que pode ser usado para realizar correspondências.
padrao = re.compile(r'\d+')

# 2. Procura a primeira ocorrência de um padrão em uma string.
match = re.search(r'\d+', 'A idade dele é 25 anos')
if match:
    print("Correspondência Encontrada:", match.group())

# 3. Verifica se o início da string corresponde ao padrão.
match_inicio = re.match(r'\d+', '1234 é um número')
if match_inicio:
    print("Correspondência no Início:", match_inicio.group())

# 4. Encontra todas as ocorrências não sobrepostas do padrão na string.
matches = re.findall(r'\d+', 'Os números são 10, 20 e 30')
print("Todas as Correspondências:", matches)

# 5. Divide a string pelo padrão de expressão regular.
dividido = re.split(r'\s+', 'Esta é uma frase com espaços')
print("String Dividida:", dividido)

# 6. Substitui as ocorrências do padrão na string por um substituto.
substituida = re.sub(r'\d+', 'NUM', 'Os valores são 10, 20 e 30')
print("String Substituída:", substituida)

# 7. Substitui as ocorrências do padrão na string por um substituto, retornando também o número de substituições feitas.
substituida, num_subs = re.subn(r'\d+', 'NUM', 'Os valores são 10, 20 e 30')
print("String Substituída:", substituida, "Número de Substituições:", num_subs)

# 8. Encontra todas as correspondências do padrão e retorna um iterador de correspondências.
for match in re.finditer(r'\d+', 'Os valores são 10, 20 e 30'):
    print("Correspondência Iterada:", match.group())

# 9. Divide a string pelo padrão de expressão regular, mas inclui o delimitador nas partes resultantes.
dividido_com_delimitador = re.split(r'(\s+)', 'Esta é uma frase com espaços')
print("String Dividida com Delimitador:", dividido_com_delimitador)

# 10. Expande um padrão de expressão regular para corresponder à string completa, incluindo os grupos nomeados e numerados.
padrao_expansivo = re.compile(r'(?P<numero>\d+)')
match_expansivo = padrao_expansivo.search('A idade dele é 25 anos')
if match_expansivo:
    print("Expansão:", match_expansivo.expand(r'Idade: \g<numero>'))

# 11. Verifica se toda a string corresponde ao padrão.
full_match = re.fullmatch(r'\d+', '123456')
if full_match:
    print("Correspondência Completa:", full_match.group())

# 12. Altera o comportamento da correspondência, tornando-a insensível a maiúsculas/minúsculas.
case_insensitive_match = re.search(r'texto', 'Este é um Texto', re.IGNORECASE)
if case_insensitive_match:
    print("Correspondência Insensível a Maiúsculas/Minúsculas:", case_insensitive_match.group())

# 13. Utiliza múltiplas linhas para correspondência, onde '^' e '$' representam o início e o fim de cada linha, respectivamente.
multiline_match = re.findall(r'^linha', 'linha 1\nlinha 2\nlinha 3', re.MULTILINE)
print("Correspondências Multilinha:", multiline_match)

# 14. Usa correspondências não-gulosas para capturar a menor quantidade possível de texto.
nongreedy_match = re.search(r'<.*?>', '<tag>conteúdo</tag>')
if nongreedy_match:
    print("Correspondência Não-Gulosa:", nongreedy_match.group())

# 15. Agrupa correspondências com parênteses para capturar partes específicas da string.
grupo_match = re.search(r'(\d+)-(\d+)', 'A faixa é 1234-5678')
if grupo_match:
    print("Grupos Capturados:", grupo_match.groups())

# 16. Nomeia grupos dentro do padrão para facilitar a referência.
grupo_nomeado_match = re.search(r'(?P<area>\d+)-(?P<local>\d+)', 'A faixa é 1234-5678')
if grupo_nomeado_match:
    print("Grupo Nomeado:", grupo_nomeado_match.group('area'), grupo_nomeado_match.group('local'))

# 17. Substitui padrões usando uma função de substituição.
def substituir_por_multiplicacao(match):
    return str(int(match.group()) * 2)

substituicao_funcional = re.sub(r'\d+', substituir_por_multiplicacao, 'Os números são 10, 20 e 30')
print("Substituição Funcional:", substituicao_funcional)
```
